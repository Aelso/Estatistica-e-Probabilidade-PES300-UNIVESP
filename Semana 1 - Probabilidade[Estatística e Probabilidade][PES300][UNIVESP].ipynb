**Fenômeno Determinístico ou Probabilístico** 🐬

**Fenômeno Determinístico** 🔶

**O resultado é sempre o mesmo**

**Fenômeno Probabilístico** 🔶
**O resultado é incerto e variável**

**Probabilidade**

**Probabilidade:** medida da informação sobre a ocorrência de um evento.

**Estatística:** Baseia-se no estudo dos fenômenos probabilísticos.

**Experimento Aleatório**

**Experimento Aleatório** pode conduzir a diferentes resultados.

Mesmo em condições iniciais iguais, o resultado é imprevisível.

**Exemplos**

* Lançamento de moedas honestas
* Lançamento de dados
* Retirada de cartas de baralhos
* Determinação da vida útil de um componente eletrônico.

**Características**

* Os possíveis resultados são conhecidos (probabilidades)
* O resultado final não é conhecido
* Cada experimento pode ser repetido indefinidamente sob as mesmas condições
* Quando o experimento for repetido um grande número de vezes e houver regularidade na explicação desse fenômeno, é possível estruturar um modelo matemático probabilístico

**Definição Frequencialista**

* m: número de sucessos
* n: número de experimentos

$lim \frac{m}{n}$

$n → ∞$

**Definição Clássica**

P: Probabilidade

M: número de sucessos

N: número de possibilidades

$P=\frac{M}{N}$

**Espaço Amostral (S)**

**Conjunto:** uma coleção definida de objetos ou itens.

**Espaço Amostral** de um experimento aleatório é o conjunto de todos os possíveis resultados desse experimento.

**Lançamento de uma moeda**

$S = {C, K}$

(é composto de 2 eventos)

**Lançamento de um dado**

$S = { 1, 2, 3, 4, 5, 6 }$

(é composto de 6 eventos)

Lançamento de 2 moedas

$S = { (C, K), (C, C), (K, C), (K, K)}$

(é composto por 4 eventos)

**Eventos Aleatórios (E)**

**Definição**

*  Qualquer subconjunto de um espaço amostral
*  Resultado possível em experimentos aleatórios e que não é previsível

**Lançamento de uma Moeda**

$S = {C,K}$

$E = K$

$P(E)=?$

**Lançamento de um dado**

$S = {1,2,3,4,5,6}$

$E = 3$

$P(E)=?$

**Lançamento de duas moedas**

$S = { (C, K), (C, C), (K, C), (K, K)}$

$E = K,K$

$P(E) = ?$

**Propriedades**

$0 ≤ $P(E) ≤ 1$

$P(S) = 1$

**Operações entre Eventos**

União - $A 𝖴 B$

image.png

Formado pelos pontos amostrais que pertencem a pelo menos um dos eventos.

Ocorre se **A** ocorre OU **B** ocorre **OU AMBOS** ocorrem.

**Intersecção - $A∩B$**

image.png

Formado pelos pontos amostrais que pertencem simultaneamente aos eventos A e B.

Ocorre apenas se **A E B** ocorrem

**Axiomas da Probabilidade** 🚧

**Se $Ø$ é o conjunto vazio (evento impossível), então $P(Ø)=0$.**
 
$P(A∩B)=0$

**2. Teorema do Evento Complementar: se $\bar{A}$ é o complemento do evento A então:**
 
$P(\bar{A}) = 1 - PA$
 
**3. Teorema da Soma: se $A\  e \ B$ são dois eventos quaisquer então:**
  
$P(A∪B) = P(A)+P(B)–P(A∩B)$

**Se $A\  e\  B$ são dois Eventos Mutuamente Excludentes**

$P(A∪B) = P(A)+P(B)$
 
**Se A, B e C são três eventos quaisquer então:**
  
$P(A∪B∪C) =P(A)+P(B)+P(C)– P(A∩B)–P(A∩C)-P(B∩C)+P(A∩B∩C)$

**Objetivos específicos da aula**

**Apresentar os principais conceitos de contagem para o cálculo de probabilidades**

**Teoria da Contagem**

**Considere dois eventos**

**Evento 1**: lançamento de uma moeda - 2 resultados possíveis

**Evento 2**: lançamento de um dado – 6 resultados possíveis

Os 2 eventos em conjunto possuem 2x6 = 12 possibilidades de ocorrerem

|		|		|		|		|		|		|
|	---------	|	---------	|	---------	|	---------	|	---------	|	---------	|
|	K,1	|	K,2	|	K,3	|	K,4	|	K,5	|	K,6	|
|	C,1	|	C,2	|	C,3	|	C,4	|	C,5	|	C,6	|


**Generalizando:**

**Evento 1**: m possibilidades

**Evento 2**: n possibilidades

**Ambos os eventos:** $m.n$ possibilidades

Para problemas **simples** ou com **poucos** elementos, pode-se contar o número de resultados de forma direta.
Para problemas **complexos**, recorre-se às **combinações** e **arranjos** para determinar o número de casos.

**Combinação**

**n**: no de elementos no conjunto amostral

**p**: número de elementos escolhidos

**c**: número de combinações possíveis a ordem dos elementos **não** faz diferença

$C_{n,p}=(\frac{n}{p})= \frac{n!}{p!n-p!}$

Exemplo: quantas comissões de três pessoas podem ser formadas com um grupo de dez pessoas?

$C_{10,3}=(\frac{10}{3})= \frac{10!}{3!10-3!}$

$=\frac{10x9x8x7!}{3x2x1(7)!} = \frac{720}{6}=120$

**Arranjos**

**r**: no de elementos no conjunto amostral

**p**: número de elementos escolhidos

**a**: número de arranjos possíveis a ordem dos elementos **faz** diferença
