**Correla√ß√£o Linear**üëå

Coeficiente de Correla√ß√£o de Pearson (r)

$r =\frac{S_{xy}}{\sqrt{S_{xx} S_{yy}}}$ ; em que

$S_{xx} ={Œ£^n _{i_{=1}} (x_{i}-\bar{x})^ 2}$

$S_{yy} ={Œ£^n _{i_{=1}} (y_{i}-\bar{y})^ 2}$



$S_{xy} ={Œ£^n _{i_{=1}} (x_{i}-\bar{x}) (y_{i}-\bar{y})}$

$S_{xx} ={Œ£x^2 _i} - \frac{({Œ£x_i})^ 2} {n}$

$S_{yy} ={Œ£y^2 _i} - \frac{({Œ£y_i})^ 2} {n}$

$S_{xy}=Œ£{x_i}.{y_i}-\frac{{Œ£x_i}.{Œ£y_i}} {n}$

Coeficiente de Correla√ß√£o de Pearson (r)

‚òë Se r pr√≥ximo de +1, existe forte correla√ß√£o positiva

‚òë Se r pr√≥ximo de -1, existe forte correla√ß√£o negativa

**Teste para o Coeficiente de Correla√ß√£o**

`M√©todo Estat√≠stico`

1. Determinar o **par√¢metro** de interesse (m√©dia, propor√ß√£o ou vari√¢ncia).
2. Definir a hip√≥tese a ser testada **H0**.
3. Definir a hip√≥tese alternativa **H1**.
4. Calcular o valor a ser comparado utilizando a **estat√≠stica de teste** apropriada.
5. Delimitar a **regi√£o de rejei√ß√£o**, valor cr√≠tico para comparar o valor anterior.
6. **Comparar** os valores obtidos.
7. **Decidir** pela rejei√ß√£o ou aceita√ß√£o da hip√≥tese H0.
8. **Concluir**, baseando-se na aceita√ß√£o ou rejei√ß√£o da hip√≥tese H0.

Para um dado **`valor de r`**, estimador do coeficiente de **`correla√ß√£o populacional œÅ`** , √© poss√≠vel concluir, a um dado **n√≠vel de signific√¢ncia Œ±**, se realmente **`existe regress√£o linear`** entre as vari√°veis.

Verificar se podemos, ao n√≠vel de confian√ßa de 95%, concluir a exist√™ncia de

**`M√©todo dos M√≠nimos Quadrados`**

Equa√ß√£o que relaciona as vari√°veis de interesse

$y =f(x_1,x_2,....,x_n) + œµ$

**`Regress√£o Linear Simples`**

$y_i = Œ± +\beta_{xi} + Œµ_i$

**`Reta Te√≥rica`**

$y_i = Œ± +\beta_{x} $

|$\hat{y}=$ | $ a +         $ |$ bx $              |
|---------- |------------------|-------------------|
|           |Coeficiente Linear|Coeficiente Angular|

**`M√©todo dos M√≠nimos Quadrados`**

A **`reta de regress√£o`** √© aquela que **`minimiza`** a soma dos quadrados das **`dist√¢ncias`** **$d_i$** (dist√¢ncias da reta aos pontos experimentais)

$min \Sigma d^2_i = min \Sigma(y_i- \hat{y}_i)^ 2 = min \Sigma (y_i- a-bx_{i})^ 2 $


$b=\frac{S_{xy}}{S_{xx}}$ onde $S_{xy}=\Sigma x_{i} y{i} - \frac{\Sigma x_{i} \Sigma y{i}}{n}$ e $S_{xx}=\Sigma x_{i}^2 - \frac{\Sigma x_{i}^2}{n}$


$a=\bar{y}-b\bar{x}$ onde $\bar{x} =\frac{\Sigma x_{i}}{n}$ e
$\bar{y}=\frac{\Sigma y_{i}}{n}$
